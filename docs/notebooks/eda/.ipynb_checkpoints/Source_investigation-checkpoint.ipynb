```python
from astropy.io import fits
hdulist_24=fits.open('/Users/pdh21/Work/Astro/dmu_products/dmu26/data/Lockman-SWIRE/MIPS/wp4_lockman-swire_mips24_map_v1.0.fits.gz')
hdulist_100=fits.open('/Users/pdh21/Work/Astro/dmu_products/dmu26/data/Lockman-SWIRE/PACS/Lockman-SWIRE_PACS100_20160921_img_gls.fits')
hdulist_160=fits.open('/Users/pdh21/Work/Astro/dmu_products/dmu26/data/Lockman-SWIRE/PACS/Lockman-SWIRE_PACS160_20160921_img_gls.fits')
hdulist_250=fits.open('/Users/pdh21/Work/Astro/dmu_products/dmu26/data/Lockman-SWIRE/SPIRE/Lockman-NEST_image_250_SMAP_v6.0.fits')
hdulist_350=fits.open('/Users/pdh21/Work/Astro/dmu_products/dmu26/data/Lockman-SWIRE/SPIRE/Lockman-NEST_image_350_SMAP_v6.0.fits')
hdulist_500=fits.open('/Users/pdh21/Work/Astro/dmu_products/dmu26/data/Lockman-SWIRE/SPIRE/Lockman-NEST_image_500_SMAP_v6.0.fits')
hdulist_lofar=fits.open('./data/lockman_full_ampphase1m_shift.app_crop.fits')

import warnings
from matplotlib.cbook import MatplotlibDeprecationWarning
warnings.simplefilter('ignore', MatplotlibDeprecationWarning)
warnings.simplefilter('ignore', UserWarning)
warnings.simplefilter('ignore', RuntimeWarning)
warnings.simplefilter('ignore',UnicodeWarning)
import aplpy
import numpy as np
```

```python
#Set some color info
import seaborn as sns
import pylab as plt
%matplotlib inline
cmap=sns.cubehelix_palette(8, start=.5, rot=-.75,as_cmap=True)
vmin=-0.01
vmax=0.1

def MRR_sample(ra,dec,radius):
    sns.set_style("white")
    bands=[]
    fig = plt.figure(figsize=(70,10))
    bands.append(aplpy.FITSFigure(hdulist_24[1],figure=fig,subplot=(1,7,1)))
    bands.append(aplpy.FITSFigure(hdulist_100[0],figure=fig,subplot=(1,7,2)))
    bands.append(aplpy.FITSFigure(hdulist_160[0],figure=fig,subplot=(1,7,3)))
    bands.append(aplpy.FITSFigure(hdulist_250[1],figure=fig,subplot=(1,7,4)))
    bands.append(aplpy.FITSFigure(hdulist_350[1],figure=fig,subplot=(1,7,5)))
    bands.append(aplpy.FITSFigure(hdulist_500[1],figure=fig,subplot=(1,7,6)))
    bands.append(aplpy.FITSFigure(hdulist_lofar[0],figure=fig,subplot=(1,7,7)))

    for i in bands:
        i.show_colorscale(cmap=sns.cubehelix_palette(8, start=.5, rot=-.75,as_cmap=True))
        i.show_markers(xid24['RA'], xid24['Dec'], edgecolor='black', facecolor='black',
                marker='o', s=100, alpha=0.5)
        i.show_markers(ra,dec, edgecolor='black', facecolor='black',
                marker='x', s=1000, alpha=1.0)
        i.show_markers(scat250['RA'],scat250['Dec'], edgecolor='blue', facecolor='blue',
    marker='o', s=100, alpha=0.5)
        i.show_markers(scat350['RA'],scat350['Dec'], edgecolor='green', facecolor='green',
    marker='o', s=100, alpha=0.5)
        i.show_markers(scat500['RA'],scat500['Dec'], edgecolor='red', facecolor='red',
    marker='o', s=100, alpha=0.5)
        i.recenter(ra, dec, width=radius*2,height=radius*2)
        i.add_colorbar()
        i.colorbar.set_location('top')
```
